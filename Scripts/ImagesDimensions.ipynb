```python
from PIL import Image
from os import listdir
import matplotlib.pyplot as plt
```

```python
Dir = './Data/Landscapes/img_landscape/'
```

```python
Files = listdir(Dir)
Sizes = []
for File in Files:
    with Image.open(Dir+File) as img:
        Sizes.append(img.size)
```

```python
Width = []
Height = []
for i in range(len(Sizes)):
    Width.append(Sizes[i][0])
    Height.append(Sizes[i][1])
```

```python
plt.hist(Width, bins=50)
plt.xlabel('Width in pixel')
plt.ylabel('Number of images')
plt.title('Histogram of the different widths of the images of the landscape dataset')
plt.show()
```

```python
plt.hist(Height, bins=50)
plt.xlabel('Height in pixel')
plt.ylabel('Number of images')
plt.title('Histogram of the different heights of the images of the landscape dataset')
plt.show()
```
